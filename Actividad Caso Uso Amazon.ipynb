### ¿Cuáles son las características principales de la mayoría de nuestros clientes (Edad, Sexo)?

### ¿Cuáles son las categorías de productos más populares?

```python
# Agrupar por "Purchase_Categories" y sumar la "Purchase_Frequency" para cada categoría
categorias_populares = df['Purchase_Categories'].value_counts()
```

### ¿Cuáles son los principales factores que influencian la frecuencia de compra?

```python
frecuencia_Compra = df['Purchase_Frequency'].value_counts()
print(frecuencia_Compra)
```

```python
multiple_times_a_week = df.loc[df['Purchase_Frequency'] == 'Multiple times a week']
```

```python
edad_frecuentes = multiple_times_a_week['age'].describe()
```

```python
genero_frecuentes = multiple_times_a_week['Gender'].value_counts()
```

```python
print("Estadísticas de la edad:\n", edad_frecuentes)
print("\nCuenta de cada valor en la columna 'Gender':\n", genero_frecuentes)
```

```python
print(df.columns)
```

```python
#
feautures_deseadas = ['Purchase_Categories', 'Personalized_Recommendation_Frequency', 'Browsing_Frequency', 'Product_Search_Method', 'Search_Result_Exploration', 'Customer_Reviews_Importance']
feautures_seleccionados = multiple_times_a_week.filter(items=feautures_deseadas)
```

```python
display(feautures_seleccionados)
```

```python
#Graficar distintas variables que puedan afectar este comportamiento
plt.figure(figsize=(8, 6))
feautures_seleccionados['Browsing_Frequency'].value_counts().plot(kind='bar')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

### ¿Qué tan satisfechos están los clientes en general con la experiencia de compra?

```python
estadisticas = df['Shopping_Satisfaction'].describe()
print(estadisticas)
```

```python
# Obtener la frecuencia de cada calificación (valor único) en la columna 'Shopping Satisfaction'
frecuencia_calificaciones = df['Shopping_Satisfaction'].value_counts()

# Crear el gráfico de barras para mostrar la distribución de las calificaciones
plt.figure(figsize=(8, 6))
plt.bar(frecuencia_calificaciones.index, frecuencia_calificaciones.values)
plt.xlabel('Calificación')
plt.ylabel('Frecuencia')
plt.title('Distribución de Calificaciones de Satisfacción de Compra')
plt.xticks(frecuencia_calificaciones.index)
plt.tight_layout()
plt.show()
```

### ¿Cuál es nuestra principal área de oportunidad de acuerdo con nuestros clientes?

```python
areas_oportunidad = df['Improvement_Areas'].value_counts()
print("Principales areas de oportunidad:", areas_oportunidad)
```

```python
top_areas_oportunidad = areas_oportunidad.head(5)
# Crear el gráfico de pastel
plt.figure(figsize=(8, 8))
plt.pie(top_areas_oportunidad, labels=top_areas_oportunidad.index, autopct='%1.1f%%', startangle=140)
plt.title('Principales Áreas de Oportunidad')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
